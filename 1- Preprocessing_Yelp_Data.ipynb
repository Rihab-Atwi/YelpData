## Read the first 100000 rows from review dataset

```python
import pandas as pd
import zipfile

# Define the path to the ZIP archive
zip_path = 'C:\\Users\\User\\Desktop\\Data yelp\\archive.zip'

# Define the name of the JSON file
json_filename = 'yelp_academic_dataset_review.json'

with zipfile.ZipFile(zip_path, 'r') as archive:
    with archive.open(json_filename) as file:
        # Read the JSON data from the ZIP archive
        df_review = pd.read_json(file, lines=True, nrows=100000)

# Display memory usage and shape
print(df_review.memory_usage(deep=True).sum())
print(df_review.shape)

df_review.head()
```

## Read business dataset

```python
import pandas as pd
import zipfile

# Define the path to the ZIP archive
zip_path = 'C:\\Users\\User\\Desktop\\Data yelp\\archive.zip'

# Define the name of the JSON file 
json_filename = 'yelp_academic_dataset_business.json'

with zipfile.ZipFile(zip_path, 'r') as archive:
    with archive.open(json_filename) as file:
        # Read the JSON data from the ZIP archive
        df_business = pd.read_json(file, lines=True)

df_business.head()
```

## Wen want only to analyze restaurant

```python
df_business = df_business[df_business['categories'].notnull()]
df_rest = df_business[df_business['categories'].str.contains('Restaurant')]
```

```python
len(df_rest)
```

```python
df_rest.head()
```

## We want to merge the two datasets to analyse the restaurant reviews. "business_id" is the common columm of the two datasets

```python
df_merged = df_review.merge(df_rest, how='inner', on='business_id')
```

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
len(df_merged)
```

```python
columns_to_drop = ['review_id', 'user_id', 'stars_x', 'useful', 'funny', 'cool', 'text', 'date']
df_merged= df_merged.drop(columns=columns_to_drop)

# Keep only unique rows based on "business_id"
df_merged= df_merged.drop_duplicates(subset='business_id')

# Display the resulting DataFrame
print(df_merged)
```

## Same steps for users 

```python
import pandas as pd
import zipfile

# Define the path to the ZIP archive
zip_path = 'C:\\Users\\User\\Desktop\\Data yelp\\archive.zip'

# Define the name of the JSON file 
json_filename = 'yelp_academic_dataset_user.json'

with zipfile.ZipFile(zip_path, 'r') as archive:
    with archive.open(json_filename) as file:
        # Read the JSON data from the ZIP archive
        df_user = pd.read_json(file, lines=True)

df_user.head()
```

```python
df_merged = df_review.merge(df_user, how='inner', on='user_id')
```

```python
columns_to_drop = ['review_id', 'business_id', 'stars', 'useful_x', 'funny_x', 'cool_x', 'text', 'date']
df_merged= df_merged.drop(columns=columns_to_drop)

# Keep only unique rows based on "business_id"
df_merged= df_merged.drop_duplicates(subset='user_id')

# Display the resulting DataFrame
print(df_merged)
```
